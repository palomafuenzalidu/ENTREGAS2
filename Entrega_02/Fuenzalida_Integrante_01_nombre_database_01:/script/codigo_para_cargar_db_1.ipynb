```python
import pandas as pd
df = pd.read_csv("BASE_DE_DATOS.csv", on_bad_lines='skip')
df.head()
```

```python
df = df.dropna(axis=1, how='all')
df = df.dropna(axis=0, how='all')
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df = df.fillna("Desconocido")
df.head()
```

```python
df.to_csv("BASE_DE_DATOS_LIMPIA.csv", index=False)
```
